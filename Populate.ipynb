```python
#modules
import pandas as pd

#import data
#Load data into python
df = pd.read_csv("bristol-air-quality-data.csv", sep = ";")

df.head()
```

```python
#writing thr Site ID 
#create a dictionary of site Ids and corresponding location
siteId = {188 : 'AURN Bristol Centre',
203 : 'Brislington Depot',
206 : 'Rupert Street',
209 : 'IKEA M32',
213 : 'Old Market',
215 : 'Parson Street School',
228 : 'Temple Meads Station',
270 : 'Wells Road',
271 : 'Trailer Portway P&R',
375 : 'Newfoundland Road Police Station',
395 : "Shiner's Garage",
452 : 'AURN St Pauls',
447 : 'Bath Road',
459 : 'Cheltenham Road \\ Station Road',
463 : 'Fishponds Road',
481 : 'CREATE Centre Roof',
500 : 'Temple Way',
501 : 'Colston Avenue'}
```

```python
a = pd.DataFrame(df.groupby(["SiteID", "geo_point_2d"]).size()).reset_index()

sites_geolocation = a[a[0] > 1]

sites_geolocation

#Split the geopoint into lat and longtitude
sites_geolocation[["Latitude", "Longitude"]] = sites_geolocation["geo_point_2d"].str.split(",", expand = True)

ids = sites_geolocation["SiteID"]
#Add the locations to the sites
location_list = []

for i in ids:
    #print(siteId.get(i))
    location_list.append(siteId.get(i))
    

sites_geolocation["Location"] = location_list

sites_geolocation[["SiteID", "Location" ,"Latitude", "Longitude"]]
```

```python
#Extract only needed columns
sitesnew = sites_geolocation[["SiteID", "Location", "Longitude", "Latitude"]]

#write data
for index, row in sitesnew.iterrows():
    cursor.execute("INSERT INTO Location (SiteId, Location, Longitude, Latitude) values(?,?,?,?)", row.SiteID, row.Location, row.Longitude, row.Latitude)

cursor.commit()
cursor.close()
```
